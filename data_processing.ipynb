```python
from torch_geometric_signed_directed.data import load_directed_real_data
from torch_geometric.datasets import Planetoid
from flepe.utils import flow_laplacian, eigen_decomposition
import torch 
```

```python
cora_data = Planetoid(root='data', name='Cora')
cora_data = cora_data[0]
_, cora_normlf_true = flow_laplacian(cora_data.edge_index, 'RGE', True)
_, cora_normlf_false = flow_laplacian(cora_data.edge_index, 'RGE', False)
cora_true_flepe = eigen_decomposition(cora_normlf_true)
cora_false_flepe = eigen_decomposition(cora_normlf_false)
torch.save(cora_true_flepe, 'PE_data/FLEPE/cora_true.pt')
torch.save(cora_false_flepe, 'PE_data/FLEPE/cora_false.pt')
```
